```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Define the input data: hours of study reshaped as a column vector
advertising_investment = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50]).reshape(-1, 1)

# Define the target data: corresponding exam scores
sales_growth = np.array([25, 30, 40, 45, 50, 60, 65, 70, 80])

# Create a LinearRegression model object
model = LinearRegression()

# Fit the model to the data (learn the best-fit line)
model.fit(advertising_investment, sales_growth)

# Print the slope (coefficient of x)
print(f"Slope (m): {model.coef_[0]:.2f}")

# Print the intercept (where the line crosses the y-axis)
print(f"Intercept (b): {model.intercept_:.2f}")

# Format the equation of the regression line as a string
equation = f"y = {model.coef_[0]:.2f}x + {model.intercept_:.2f}"
print(f"Line equation: {equation}")


# Plotting section
plt.figure(figsize=(10, 10))  # Set the figure size

# Plot the original data points as blue dots
plt.scatter(advertising_investment, sales_growth, color='fuchsia', label='Data points')

# Plot the regression line (predicted scores)
plt.plot(advertising_investment, model.predict(advertising_investment), color='pink', label='Regression line')

# Mark the prediction point (for 60 אלף ש"ח) in blue
score_to_predict = 60
invest_60_needed = model.predict([[score_to_predict]])[0]
# Mark the prediction point (for 100 אלף ש"ח) in blue
for_100_needed = (100 - model.intercept_) / model.coef_[0]

plt.scatter([[for_100_needed]], [100], color='blue', s=100, label='Our prediction')

# Add chart title and axis labels
plt.title('Linear Regression - Advertising Investment vs. Sales Growth')
plt.xlabel('Investment in advertising')
plt.ylabel('sales growth')
plt.grid(True)  # Add grid to the background
plt.legend()    # Show legend

# Display the equation of the line on the plot
plt.text(10, 80, equation, fontsize=15)

# Show the final plot
plt.show()
```

```python
# Answer to the question 2
score_to_predict = 60_000 
# x = y - b / m
invest_60_needed = model.predict([[score_to_predict]])[0]
print(f"If the company invests 60 thousand shekels in advertising, it will earn from sales: {invest_60_needed:.2f}")
```

```python
# Answer to the question 3
score_to_predict = 100_000 
 # x = y - b / m
for_100_needed = (score_to_predict - model.intercept_) / model.coef_[0]
print(f"The company needs to invest in advertising: {for_100_needed:.2f} in order to see an increase of 100,000 shekels in sales")
```
